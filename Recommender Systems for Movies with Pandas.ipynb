# Recommending Similar Movies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('white')
```

```python
#Get the data
#This data is a subset of : MovieLens - Movie Recommendation Data Sets http://www.grouplens.org/node/73

column_names = ['user_id', 'item_id', 'rating', 'timestamp']
df = pd.read_csv('u.data', sep = '\t', names = column_names)
```

```python
#bring movie titles
movie_titles = pd.read_csv('Movie_Id_Titles')
movie_titles.head()
```

```python
#merge item id and movie title
df = pd.merge(df, movie_titles, on = 'item_id')
```

```python
df.head()
```

```python
#mean rating for every movie
df.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
#movies with the most ratings
df.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
```

```python
ratings['num of ratings']= pd.DataFrame(df.groupby('title')['rating'].count())
```

```python
ratings.head()
```

```python
sns.histplot(ratings['rating'], palette = 'pastel', bins = 70);
```

## Building a Recommender System for the Movie Data

```python
#converting dataframe into matrix
moviemat = df.pivot_table(index = 'user_id', columns = 'title', values = 'rating')
```

```python
moviemat
```

```python
ratings.sort_values('num of ratings', ascending = False).head(10)
```

```python
starwars_user_ratings = moviemat['Star Wars (1977)']
liarliar_user_ratings = moviemat['Liar Liar (1997)']
```

```python
starwars_user_ratings.head()
```

```python
liarliar_user_ratings.head()
```

```python
similar_to_starwars = moviemat.corrwith(starwars_user_ratings)
similar_to_starwars
```

```python
similar_to_liarliar = moviemat.corrwith(liarliar_user_ratings)
similar_to_liarliar
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars, columns = ['Correlation'])
corr_starwars.dropna(inplace = True)
```

```python
corr_starwars.head()
```

```python
corr_starwars.sort_values('Correlation', ascending = False).head(10)
```

### We get movies with perfect correlation with Star Wars. Most likely these movies happened to be seen by one person, who also happened to rate star wars 5 stars. We can fix this by filterting out certain movies that have less than a certain number of reviews, setting a threshold of num of ratings necessary.

```python
corr_starwars = corr_starwars.join(ratings['num of ratings'])
```

```python
corr_starwars.head()
```

```python
#drop all with num of ratings less than 100
corr_starwars = corr_starwars[corr_starwars['num of ratings']>100].sort_values('Correlation', ascending = False)
corr_starwars.head()
```

```python
corr_liarliar = pd.DataFrame(similar_to_liarliar, columns = ['Correlation'])
corr_liarliar.dropna(inplace = True)
```

```python
corr_liarliar = corr_liarliar.join(ratings['num of ratings'])
corr_liarliar = corr_liarliar[corr_liarliar['num of ratings']>100].sort_values('Correlation', ascending = False)
corr_liarliar.head()
```
